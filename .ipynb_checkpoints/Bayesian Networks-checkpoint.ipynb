```python
import pandas as pd
import time
import calendar
```

```python
def weekdayColumn(df):
    '''
    Takes in pandas dataframe and creates 'weekday' column which represents
    the day of the week corresponding to Date.
    '''
    dates = [str(d).replace('/','') for d in list(df['Date'])]
    weekdays = [time.strptime(d,'%m%d%Y').tm_wday for d in dates]
    weekdays = [calendar.day_name[w] for w in weekdays]
    df['weekday'] = weekdays
    
def dateColumns(df):
    dates = [str(d).replace('/','') for d in list(df['Date'])]
    dates = [time.strptime(d,'%m%d%Y') for d in dates]
    df['MONTH'] = [d.tm_mon for d in dates]
    df['DAY'] = [d.tm_mday for d in dates]

def monthNameColumn(df):
    '''
    Takes in pandas dataframe and creates 'month' column.
    '''
    dates = [str(d).replace('/','') for d in list(flights_df['Date'])]
    dates = [time.strptime(d,'%m%d%Y').tm_mon for d in dates]
    df['month'] = [calendar.month_name[m] for m in dates]
    
def filterRoutes(df):
    '''
    filter df to contain delayed fligths from SFO to PHL routes only.
    '''
    df = df[(df['ORIGIN'] == 'SFO') & (df['DEST'] = 'PHL') & (df['ARR_DELAY'] > 0)]
    df = df.fillna(0, inplace=True)
```

Read in Airport information from ASPM (SFO-PHL pair city data and individual airport data files) and merge into flights_df.

```python
# read in SFO to PHL flight delay information
dot_df = pd.read_csv('flights.csv').dropna(axis=0,how='all')
# keep departure, arrival hour info, drop minutes
dot_df['SCHEDULED_DEPARTURE'] = [t // 100 for t in list(dot_df['SCHEDULED_DEPARTURE'])]
dot_df['SCHEDULED_ARRIVAL'] = [t // 100 for t in list(dot_df['SCHEDULED_ARRIVAL'])]
# remove ARRIVAL_DELAY = 0, CANCELLED != 0
dot_df = dot_df[(dot_df['ORIGIN_AIRPORT']=='SFO') & (dot_df['DESTINATION_AIRPORT']=='PHL') & (dot_df['ARRIVAL_DELAY'] != 0.0) & (dot_df['CANCELLED'] == 0)].reset_index()
dot_df = dot_df.drop(['index','YEAR','FLIGHT_NUMBER','ORIGIN_AIRPORT','DESTINATION_AIRPORT','DEPARTURE_TIME','ARRIVAL_TIME','TAIL_NUMBER','CANCELLED','CANCELLATION_REASON'],axis=1)
dot_df.fillna(0,inplace=True)
```

```python
# read in detailed airport information

airports_df = pd.read_csv('airport_delay.csv').dropna(axis=0,how='all')
# split Date into MONTH and DAY
airports_df['Facility'] = airports_df['Facility'].str.strip()
dateColumns(airports_df)
airports_df = airports_df.drop(['% DelayedGateDepartures','% DelayedGateArrivals','Date'],axis=1)
airports_df = airports_df.drop(['DeparturesFor MetricComputation','ArrivalsFor MetricComputation','DelayedGateDepartures','DelayedGateArrivals','AverageMinutesOf DelayPer DelayedGateDeparture','AverageMinutesOf DelayPer DelayedGateArrival'],axis=1)

# measure of congestion
airports_df['netOutfluxPHL'] = airports_df['ScheduledDepartures'] - airports_df['ScheduledArrivals']
airports_df['netInfluxSFO'] = airports_df['ScheduledArrivals'] - airports_df['ScheduledDepartures']
airports_df = airports_df.rename(columns={'ScheduledDepartures':'ScheduledDeparturesCount',\
                                         'ScheduledArrivals':'ScheduledArrivalsCount'})
```
